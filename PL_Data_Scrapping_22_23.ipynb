```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://fbref.com/en/comps/9/2022-2023/2022-2023-Premier-League-Stats'
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find('table')
```

```python
table_heads = table.find_all('th', class_ ="poptip")
```

```python
column = [head.text.strip() for head in table_heads]

column = column[0:-1]
```

```python
df = pd.DataFrame(columns=column)

df
```

```python
table_data = table.find_all('tr')
```

```python
for data in table_data[1:]:
    row_data = data.find_all('td')
    ind_data = [rdata.text.strip() for rdata in row_data]
    team = data.find('th')
    ind_data = [team, *ind_data[0:-1]]
    length = len(df)
    df.loc[length] = ind_data
    
df
```

```python
import numpy as np

df['Rk'] = pd.Series(np.arange(1,21))
```

```python
df.columns
```

```python
df['Attendance'] = df['Attendance'].apply(lambda x: x.replace(',', ''))
```

```python
to_num = ['Rk', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 'Pts/MP', 'xG', 'xGA', 'xGD', 'xGD/90', 'Attendance']

df[to_num] = df[to_num].apply(pd.to_numeric)
```

```python
df
```

```python
df.to_csv(r'C:\Users\Owner\OneDrive\Desktop\Python\Arsenal Data Scraping\PL_Data_22_23.csv', index=False)
```

```python
url = 'https://fbref.com/en/comps/9/2022-2023/schedule/2022-2023-Premier-League-Scores-and-Fixtures'
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find(id='sched_2022-2023_9_1')
```

```python
table_heads = table.find_all('th', class_ ="poptip")
```

```python
column = [head.text.strip() for head in table_heads]

column = column[3:-2]
```

```python
df = pd.DataFrame(columns=column)

df
```

```python
table_data = table.find_all('tr')
```

```python
for data in table_data[1:]:
    row_data = data.find_all('td')
    ind_data = [rdata.text.strip() for rdata in row_data]
    ind_data = ind_data[2:-2]
    
    length = len(df)
    df.loc[length] = ind_data
    
df
```

```python
df.info()
```

```python
df = df.drop(df[df['Score'] == ''].index)
```

```python
df[['Home_Goals', 'Away_Goals']] = df['Score'].str.split('–', expand=True)
```

```python
df['Attendance'] = df['Attendance'].apply(lambda x: x.replace(',', ''))
```

```python
to_num = ['Attendance', 'Home_Goals', 'Away_Goals']

df['xG'] = df['xG'].apply(pd.to_numeric)
df[to_num] = df[to_num].apply(pd.to_numeric)
df = df.drop(columns=['Score'])
```
