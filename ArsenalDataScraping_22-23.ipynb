```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://fbref.com/en/squads/18bb7c10/2022-2023/Arsenal-Stats'
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
standard_data_table = soup.find(id='stats_standard_9')
```

```python
standard_table_headers = standard_data_table.find_all('th', class_='poptip')
```

```python
column = [headers.text.strip() for headers in standard_table_headers]

column = column[0:-11]
```

```python
df = pd.DataFrame(columns=column)

df
```

```python
standard_body = standard_data_table.find('tbody')
standard_data = standard_body.find_all('tr')

standard_data
```

```python
for data in standard_data:
    row_data = data.find_all('td')
    ind_data = [dt.text.strip() for dt in data]
    ind_data = ind_data[0:-11]
    length = len(df)
    df.loc[length] = ind_data
    
df
```

```python
df= df.drop(df[(df['MP'] == '0') & (df['Starts'] == '0')].index)
```

```python
df.info()
```

```python
df.columns
```

```python
my_cols = ['MP', 'Starts', 'Min', '90s', 'Gls',
       'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'xG', 'npxG',
       'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR']

df[my_cols] = df[my_cols].apply(lambda x: x.str.replace(',', ''))
```

```python
to_num = ['90s', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'MP', 'Starts', 'Min', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'PrgC', 'PrgP', 'PrgR']


df[to_num] = df[to_num].apply(pd.to_numeric)

df.info()
```

```python
df.to_csv(r'C:\Users\Owner\OneDrive\Desktop\Python\Arsenal Data Scraping\arsenal_standard_data_22-23.csv', index=False)
```

```python
fix_url = 'https://fbref.com/en/squads/18bb7c10/2022-2023/matchlogs/c9/schedule/Arsenal-Scores-and-Fixtures-Premier-League'
```

```python
page = requests.get(fix_url)

soup = BeautifulSoup(page.text, 'html')
```

```python
fix_table = soup.find('table')
```

```python
fix_data_header = fix_table.find_all('th', class_='poptip')
```

```python
column = [col.text.strip() for col in fix_data_header]

column = column[0:-2]
```

```python
df = pd.DataFrame(columns=column)

df
```

```python
fix_table_body = fix_table.find('tbody')
fix_body_rows =fix_table_body.find_all('tr')
```

```python
for data in fix_body_rows:
    row_data = data.find_all('td')
    ind_data = [dt.text.strip() for dt in row_data]
    ind_data = ind_data[0:-2]
    date_of_match = data.find('th').text.strip()
    length = len(df)
    df.loc[length] = [date_of_match, *ind_data]
    
df
```

```python
df.info()
```

```python
df.columns
```

```python
df[['GF', 'GA', 'Poss', 'Attendance']] = df[['GF', 'GA', 'Poss', 'Attendance']].apply(lambda x: x.str.replace(',', ''))
df['Round'] = df['Round'].str.replace('Matchweek ', '')
```

```python
to_float = ['xG', 'xGA']
to_int = ['GF', 'GA', 'Poss', 'Attendance', 'Round']

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df[to_float] = df[to_float].astype(float)
df[to_int] = df[to_int].astype(int)

df.info()
```

```python
df.to_csv(r'C:\Users\Owner\OneDrive\Desktop\Python\Arsenal Data Scraping\fixtures_22_23.csv', index=False)
```
